4. what is the relationship between the sale price of iphones and the number of ratings on flipkart

```python
import plotly.express as px
```

```python
fig = px.scatter(df, x='Number Of Ratings', y='Sale Price', color='Discount Percentage',trendline='ols')
fig.show()
```

```python
# general trend is sale price tends to decrease when number of ratings increase hence negative linear relationship
```

5. what is the relationship between the discount percentage and the number of ratings of iphones on flipkart

```python
# we get to see a positive linear relationship as the discount percentge increases the number of ratings increases
```
